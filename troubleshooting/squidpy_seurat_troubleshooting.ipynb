# Squidpy + Seurat Troubleshooting Roundup

Real errors hit during the Squidpy vs Seurat comparison, and how we fixed them.

Blog post: [spatiabio.com](https://www.spatiabio.com)

## Error 1: `GetAssayData()` with `slot` is defunct (Seurat v5)

```python
# Wrong (Seurat v4 syntax)
# data.input <- GetAssayData(brain, slot = "data", assay = "SCT")

# Fixed (Seurat v5)
# data.input <- GetAssayData(brain, layer = "data", assay = "SCT")
print("Use 'layer=' instead of 'slot=' in Seurat v5")
```

## Error 2: `highly_variable_genes` with seurat_v3 flavor requires scikit-misc

```python
# Wrong - requires skmisc package
# sc.pp.highly_variable_genes(adata, n_top_genes=2000, flavor="seurat_v3")

# Fixed - use cell_ranger flavor instead
import scanpy as sc
import squidpy as sq
import warnings; warnings.filterwarnings("ignore")

adata = sq.datasets.visium_hne_adata()
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, n_top_genes=2000, flavor="cell_ranger")
print(f"HVGs found: {adata.var.highly_variable.sum()}")
```
