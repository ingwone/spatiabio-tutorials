# Spatio-DARLIN: Spatial Lineage Tracing

Code companion for the Spatio-DARLIN paper review.

Paper: [Nature Methods, 2026](https://www.nature.com/nmeth/)

Blog post: [spatiabio.com](https://www.spatiabio.com)

## What Spatio-DARLIN does

Spatio-DARLIN combines CRISPR-based lineage barcodes with spatial transcriptomics to reconstruct **where cells have been** as well as their current location.

Key outputs:
- Clonal barcode assignments per cell
- Spatial clone maps
- Lineage trees with spatial context

## Conceptual simulation

The actual Spatio-DARLIN pipeline requires proprietary barcode data. Below is a conceptual demonstration using public Visium data to illustrate spatial clonal analysis.

```python
import squidpy as sq
import scanpy as sc
import numpy as np
import warnings; warnings.filterwarnings("ignore")

adata = sq.datasets.visium_hne_adata()
sc.pp.normalize_total(adata, target_sum=1e4)
sc.pp.log1p(adata)
sc.pp.highly_variable_genes(adata, n_top_genes=2000, flavor="cell_ranger")
sc.pp.pca(adata)
sc.pp.neighbors(adata)
sc.tl.leiden(adata, key_added="cluster")

# Simulate clone assignments (illustrative)
np.random.seed(42)
adata.obs["simulated_clone"] = np.random.choice(
    ["Clone_A", "Clone_B", "Clone_C", "Clone_D"], size=adata.n_obs
)
print(adata.obs["simulated_clone"].value_counts())
```

```python
sq.gr.spatial_neighbors(adata)
sq.pl.spatial_scatter(adata, color=["cluster", "simulated_clone"],
                      title=["Leiden clusters", "Simulated clones (illustrative)"])
```

## Key concepts from the paper

- **DARLIN barcodes**: Evolving CRISPR scars recorded across cell divisions
- **Spatial registration**: Barcodes linked to Visium spot coordinates
- **Clone reconstruction**: Shared barcode sequences = shared lineage
