# Co-occurrence Analysis with Squidpy

Companion notebook for: [spatiabio.com — post 8](https://www.spatiabio.com/2026/07/co-occurrence-analysis-squidpy.html)

Compute distance-resolved co-occurrence probability between Leiden clusters on 10x Genomics Visium mouse brain data using `sq.gr.co_occurrence`.

```python
import squidpy as sq
import scanpy as sc
import warnings
warnings.filterwarnings('ignore')
```

```python
adata = sq.datasets.visium_hne_adata()

# NOTE: visium_hne_adata() ships .X ALREADY log-normalized (max ~8.4, non-integer);
# the raw integer counts live in adata.raw.X. Calling normalize_total + log1p here
# would log it a second time -- silently, with no error, changing every result below.
sc.pp.highly_variable_genes(adata, n_top_genes=2000, flavor='cell_ranger')
sc.pp.pca(adata)
sc.pp.neighbors(adata)
sc.tl.leiden(adata, key_added='cluster')
```

```python
sq.gr.spatial_neighbors(adata)
sq.gr.co_occurrence(adata, cluster_key='cluster')
```

```python
top_clusters = adata.obs['cluster'].value_counts().head(3).index.tolist()

sq.pl.spatial_scatter(adata, color='cluster', title='Leiden clusters on tissue')

sq.pl.co_occurrence(
    adata,
    cluster_key='cluster',
    clusters=top_clusters,
    figsize=(10, 4)
)
```
